## Data Frame

a data structure native to pandas, similar to excel but has lots of functions: lot of ways to access data unlike excel.

### Basic Functions of Data Frame

<pre>DataFrame.shape()                              calculate the dimensions of the DataFrame "rows X columns"
DataFrame.head(n)/DataFrame.tail(n)            Access top/bottom (n) records
DataFrame.columns                              Access all column names
DataFrame["column_name"]                       Access data of one column
DataFrame[["column1","column2",...]]           Access data of multiple columns</pre>

```python
# seeing the dimensions of the df data frame

df.shape
```

```python
# print number of rows in data frame

df.shape[0]
```

```python
# bottom 5 rows

df.tail()
```

```python
# last 10 rows

df.tail(10)
```

```python
# seeing the names of columns in df dataframe

df.columns
```

```python
# selecting a single column

df['SUBJECT']
```

```python
# selecting multiple columns using names

df[['SUBJECT','FACULTY NAME']]
```
