# Pandas

## Indexing a DataFrame

```python
# importing pandas library

import pandas as pd
```

```python
# reading the dataset

df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
# selecting rows by their position

df.iloc[:5]
```

```python
# selecting columns by their position

df.iloc[:,:2]
```

```python
df.iloc[5:10,:2]
```

```python
# selecting rows by their values

df[df['SUBJECT']=='PC']
```
